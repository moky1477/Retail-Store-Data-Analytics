```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../Data/orders.csv')
df.head()
```

# Data Cleaning

```python
# Check for unique values, and fix nan values
df["Ship Mode"].unique()
```

```python
df["Ship Mode"] = df["Ship Mode"].apply(lambda x: np.nan if x=='unknown' or x=='Not Available' else x)
```

```python
df["Ship Mode"].unique()
```

```python
df.head(20)
```

```python
# Fix column names (remove spaces)
df.columns
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
# Create a discount column
df["discount"] = df["list_price"]*(df["discount_percent"]/100)
```

```python
# Create selling price column
df["selling_price"] = df["list_price"]-df["discount"]
```

```python
# Calculate total profit column
df["total_profit"] = df["selling_price"]-df["cost_price"]
```

```python
df.head()
```

```python
# Convert order_date to date
df.dtypes
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], format="%Y-%m-%d")
```

```python
df.dtypes
```

```python
# Drop unwanted columns
df.drop(["cost_price", "list_price", "discount_percent"], axis=1, inplace=True)
```

```python
df.head()
```
